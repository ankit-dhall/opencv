Canny Edge Detection is a popular edge detection algorithm. For theory on how Canny Edge Detection works, visit https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_canny/py_canny.html

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Images/bottle.jpg',0)
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
